```python
import json
import pandas as pd
```

```python
zip_hrsa = pd.read_excel('json/ZIP_Code_to_ZCTA_Crosswalk_hrsa.xlsx')
zip_hrsa['zcta'] = zip_hrsa['zcta'].fillna(-1).astype(int).astype(str).replace('-1', '').str.zfill(5).replace('00000', '')
zip_hrsa = zip_hrsa.rename(columns={'ZIP_CODE': 'zip_code', 'zcta': 'zcta_hrsa'})
zip_hrsa['zip_code'] = zip_hrsa['zip_code'].astype(int).astype(str).str.zfill(5).replace('00000', '')
zip_hrsa
```

```python
zip_zct_xref = pd.read_csv('json/zip_zcta_xref.csv')
zip_zct_xref['zip_code'] = zip_zct_xref['zip_code'].astype(int).astype(str).str.zfill(5).replace('00000', '')
zip_zct_xref['zcta'] = zip_zct_xref['zcta'].fillna(-1).astype(int).astype(str).replace('-1', '').str.zfill(5).replace('00000', '')
```

```python
merged_df = pd.merge(zip_hrsa, zip_zct_xref, on='zip_code', how='outer')
merged_df
```

```python
merged_df[merged_df['zcta'] != merged_df['zcta_hrsa']]
```

# 2010

```python
with open("json/zipzcta_crosswalk_2010.json") as open_cross:
    # A dictionary containing the ZIP-Code (key; string)
    # and corresponding ZCTA (value; string) for the given [year].
    zipzcta_crosswalk = json.load(open_cross)
zipzcta_crosswalk = pd.DataFrame(zipzcta_crosswalk.items(), columns=['zip_code', 'zcta_crosswalk'])

zipzcta_crosswalk['zip_code'] = zipzcta_crosswalk['zip_code'].astype(int).astype(str).str.zfill(5).replace('00000', '')
zipzcta_crosswalk['zcta_crosswalk'] = zipzcta_crosswalk['zcta_crosswalk'].fillna("-1").astype(str).replace('-1', '').str.zfill(5).replace('00000', '')
zipzcta_crosswalk
```

```python
df = pd.merge(zip_zct_xref, zipzcta_crosswalk, left_on='zip_code', right_on='zip_code', how='outer')
df
```

```python
unmatched = df[df['zcta'] != df['zcta_crosswalk']]

df.to_csv('json/comparison.csv', index=False)
unmatched.to_csv('json/unmatched.csv', index=False)
```

```python
unmatched
```

```python
unmatched.dropna(subset=['zcta'], inplace=False).reset_index(drop=True)
```

```python
unmatched.dropna(subset=['zcta_crosswalk'], inplace=False).reset_index(drop=True)
```

```python
unmatched.dropna(subset=['zcta_crosswalk', 'zcta'], inplace=False).reset_index(drop=True)
```
